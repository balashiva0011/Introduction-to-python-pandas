```python
df = df[df['Country'] == 'Canada']
```

```python
df
```

```python
df = df.rename(columns={'State	': 'States', 'Sub_Category': 'Sub_Categorys'})

print(df)
```

```python
rows = df.iloc[[0, 1],[0,1]]
print(rows)
```

```python
rows = df.iloc[0]
print(rows)
```

```python
columns = df.iloc[:, 0:2]
print(columns)
```

```python
df = df.drop("Unit_Cost",axis=1)
print(df)
```

```python
result = df['Revenue'].agg(['sum','mean'])
print(result)
```

```python
result = df['Revenue'].agg(['sum','mean','max','min'])
print(result)
```

```python
result = df.agg({
    'Revenue': ['mean', 'sum'],
    'Profit': ['max', 'min']})
print(result)
```

```python
grouped_agg = df.groupby('Sub_Categorys').agg({
    'Revenue': ['mean', 'sum', 'max', 'min']})
print(grouped_agg)
```

```python
def range_func(x):
    return x.max() - x.min()

# Apply custom function to grouped data
custom_agg = df.groupby('Sub_Categorys').agg({
    'Revenue': range_func,
    'Profit': 'mean'
})

print(custom_agg)
```

```python
import matplotlib.pyplot as plt
```

```python
sales_trends = df.groupby('Sub_Categorys')['Revenue'].sum()

# Plot
plt.figure(figsize=(10, 6))
plt.plot(sales_trends.index, sales_trends.values, marker='o')
plt.title('Bike Sales Trends Over Time')
plt.xlabel('Sub_Categorys')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Revenue'], labels=df['Product'], autopct='%1.1f%%', startangle=140)
plt.title('Market Share by Customer Segment')
plt.show()
```

```python
sales_by_type = df.groupby('Sub_Categorys')['Revenue'].sum()

# Plot
plt.figure(figsize=(10, 6))
sales_by_type.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Bike Type')
plt.xlabel('Sub_Categorys')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.boxplot(column='Revenue', by='Sub_Categorys', grid=False)
plt.title('Sales Distribution by Country')
plt.suptitle('')  # Remove the default title to avoid overlap
plt.xlabel('Sub_Categorys')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
