## **A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns.**

```python
import pandas as pd
import numpy as np
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
#load data into a DataFrame object:
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
#refer to the row index:
print(df.loc[0])
```

```python
#use a list of indexes:
print(df.loc[[0, 1]])
```

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

df
```

```python
print(df.loc["day2"])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 6))  # Optional: Adjust figure size

# Scatter plot
plt.scatter(df['calories'], df['duration'], color='g', marker='o')

# Adding labels and title
plt.xlabel('Calories')
plt.ylabel('Duration')
plt.title('Calories vs Duration of Workout')

# Show plot
plt.grid(True)  # Optional: Add grid
plt.tight_layout()  # Optional: Adjust layout
plt.show()
```

```python
plt.figure(figsize=(8, 4))  # Optional: Adjust figure size


# Plotting bars
plt.bar( df['calories'],  color='b', label='Calories')
plt.bar( df['duration'],  color='r', label='Duration')

# Adding labels and title
plt.xlabel('Workout')
plt.ylabel('Value')
plt.title('Calories Burned and Duration of Workout')
plt.legend()

# Show plot
plt.grid(True)  # Optional: Add grid
plt.tight_layout()  # Optional: Adjust layout
plt.show()
```

```python
# Data for the pie chart
labels = ['Workout 1', 'Workout 2', 'Workout 3']
calories = df['calories']

# Plotting the pie chart
plt.figure(figsize=(6, 6))  # Optional: Adjust figure size
plt.pie(calories, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Distribution of Calories Burned by Workout')

# Show plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()  # Optional: Adjust layout
plt.show()
```
