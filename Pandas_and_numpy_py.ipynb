NumPy
NumPy (Numerical Python) is a library for numerical computing in Python. It provides support for arrays, matrices, and a large collection of mathematical functions to operate on these data structures.

```python
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])

# Calculate the mean
mean = np.mean(arr)
print(f"Mean: {mean}")

# Matrix multiplication
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
result = np.dot(matrix1, matrix2)
print("Matrix multiplication result:\n", result)
```

Pandas
Pandas is a library for data manipulation and analysis, providing data structures like Series and DataFrames. It is built on top of NumPy and offers powerful tools for working with structured data.

Example 2: Using the @ Operator

```python
import numpy as np

# Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Perform matrix multiplication
C = A @ B
print("Matrix C (using @ operator):\n", C)
```

Example 3: Using np.matmul()


```python
import numpy as np

# Define two matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Perform matrix multiplication
C = np.matmul(A, B)
print("Matrix C (using np.matmul):\n", C)
```

```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df = pd.DataFrame(data)

df
```

```python
# Display basic statistics
print(df.describe())

# Filter rows where Age > 28
filtered_df = df[df['Age'] > 28]
print(filtered_df)

```

```python
# Group by 'City' and calculate the mean age
grouped = df.groupby('City')['Age'].mean()
print(grouped)
```

Combining Pandas and NumPy

```python
import pandas as pd
import numpy as np

# Create a DataFrame
df = pd.DataFrame({
    'A': [1, 2, 3, np.nan, 5],
    'B': [5, 6, np.nan, 8, 9]
})
df.isnull().sum()
```

```python
df.mean()
```

```python
# Fill missing values using the mean of the column
df.fillna(df.mean(), inplace=True)
df
```
