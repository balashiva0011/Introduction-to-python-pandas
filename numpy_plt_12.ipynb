```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 34, 56])
y = np.array([2, 67, 89, 90])
```

```python
plt.bar(x, y)
plt.title('Bar Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
plt.hist(x, bins=5)  # Adjust bins as needed
plt.title('Histogram')
plt.xlabel('X-axis')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
labels = ['Label 1', 'Label 2', 'Label 3', 'Label 4']
plt.pie(y, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python

plt.scatter(x, y)
plt.title('Scatter Plot of x and y')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```
